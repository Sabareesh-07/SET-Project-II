```python
import pandas as pd
df = pd.read_csv(r"E:\SET Project II\Data\Power comsumption\household_power_consumption.txt", sep=";")
print(df.head())
```

```python
df.isnull().sum()
(df.isnull().sum() / len(df)) * 100
```

```python

#replace empty strings, "NA", "null", and "?" with pd.NA
df.replace(["", "NA", "null","?"], pd.NA, inplace=True)
df.isnull().sum()
```

```python
#drop rows with missing values  
print("Before:", df.shape)
data = df.dropna()
print("After:", data.shape)
print("percent of data lost:", (1 - (len(data) / len(df))) * 100.00)
```

```python
print(data.dtypes)
# convert columns to numeric, errors='ignore' will keep non-numeric values as is
data = data.apply(pd.to_numeric,errors='ignore')
```

```python
print(data.dtypes)
```

```python
data.head()
```

```python
#merge date and time columns to create a datetime column
data["Date"] = pd.to_datetime(data["Date"], format=r"%d/%m/%Y", errors="coerce")
data["Time"] = pd.to_datetime(data["Time"], format=r"%H:%M:%S", errors="coerce").dt.time
data["datetime"] = pd.to_datetime(
    data["Date"].astype(str) + " " + data["Time"].astype(str),
    errors="ignore"
)
print(data.dtypes)
```

```python
data.head()
```

```python
data = data.drop(['Date', 'Time'], axis=1)
data.head()
```

```python
# convert minutely data to hourly data
data = data.set_index('datetime')
data = data.sort_index()
data = data.resample('h').mean()
print(data.isnull().sum())
```

```python
# fill missing values with forward fill method
data = data.ffill()
print(data.isnull().sum())
```

```python
# extract date features
data['hour'] = data.index.hour
data['day'] = data.index.day
data['month'] = data.index.month
data['weekday'] = data.index.weekday
```

```python
# create lag features
data['lag_1'] = data['Global_active_power'].shift(1)
data['lag_2'] = data['Global_active_power'].shift(2)
data['lag_24'] = data['Global_active_power'].shift(24)
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
data = data.dropna()
print(data.info())
print(data.head())
```

```python
# create rolling mean features
data['rolling_mean_3'] = data['Global_active_power'].shift(1).rolling(3).mean()
data['rolling_mean_24'] = data['Global_active_power'].shift(1).rolling(24).mean()
data = data.dropna()
```

```python
print(data.head())
print(data.info())
```

```python
data = data.reset_index(drop=False)
data.to_csv(r"E:\SET Project II\Cleaned_data.csv", index=False)
```
